```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import  accuracy_score
```

```python
path = "https://docs.google.com/spreadsheets/d/e/2PACX-1vR3X81gX_Zlbs4j244cQbBe04CTZbJ-Dl6EwNzKZRFngs9XZu_9dzVDuzzpZJgCj6xqS8oHR8K3eqr9/pub?output=csv"
df = pd.read_csv(path, index_col=0)
df
```

```python
X = df.drop(columns='diagnosis')
y = df['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_train
```

1) clasificador del árbol de decisión

```python
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train, y_train)

# Obtain predictions
train_preds = tree_clf.predict(X_train)
test_preds = tree_clf.predict(X_test)

# Calculate classification accuracy
train_score = accuracy_score(y_train, train_preds)
test_score = accuracy_score(y_test, test_preds)
print (f'The training accuracy is {train_score:.3f}.')
print (f'The testing accuracy is {test_score:.3f}.')
```

2) clasificador bagging

```python
bag_clf = BaggingClassifier(random_state=42)
bag_clf.fit(X_train, y_train)

# Obtain predictions
train_preds = bag_clf.predict(X_train)
test_preds = bag_clf.predict(X_test)

# Calculate classification accuracy
train_score = accuracy_score(y_train, train_preds)
test_score = accuracy_score(y_test, test_preds)
print (f'The training accuracy is {train_score:.3f}.')
print (f'The testing accuracy is {test_score:.3f}.')
```

3) clasificador de bosque aleatorio

```python
rf_clf = RandomForestClassifier(random_state=42)
rf_clf.fit(X_train, y_train)

# Obtain predictions
train_preds = rf_clf.predict(X_train)
test_preds = rf_clf.predict(X_test)

# Calculate classification accuracy
train_score = accuracy_score(y_train, train_preds)
test_score = accuracy_score(y_test, test_preds)
print (f'The training accuracy is {train_score:.3f}.')
print (f'The testing accuracy is {test_score:.3f}.')
```

El accuracy más alto corresponde a Random Forest, el cual es de 0.965 en el conjunto de test mientras que los otros algoritmos como Árbol de Decisión y Bagging tienen un accuracy de 0.951 ambos respectivamente.
