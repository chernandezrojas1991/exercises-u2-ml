```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
```

```python
path = "https://docs.google.com/spreadsheets/d/e/2PACX-1vR3X81gX_Zlbs4j244cQbBe04CTZbJ-Dl6EwNzKZRFngs9XZu_9dzVDuzzpZJgCj6xqS8oHR8K3eqr9/pub?output=csv"
df = pd.read_csv(path, index_col=0)
df
```

Definir variable objetivo o target

```python
target = 'diagnosis'
X = df.drop(columns=target).copy()
y = df[target].copy()

target_map = {'M':1,'B':0}
y = y.map(target_map)
```

Definir conjunto de entrenamiento y test con train_test_split.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_train
```

FunciÃ³n de evaluaciÃ³n

```python
def evaluate_classification(model,X_train,X_test,y_train,y_test):
    """Uses model.score to display the R^2 values for training and test splits"""
    print(f"Training Accuracy: {model.score(X_train,y_train):.3f}")
    print(f"Test Accuracy: {model.score(X_test,y_test):.3f}")
```

Instanciar Ã¡rbol de decisiÃ³n y ajustar modelo con el conjunto de entrenamiento y testeo. Luego, imprimir resultados.

# Ãrbol de DecisiÃ³n

```python
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train, y_train)

evaluate_classification(tree_clf, X_train, X_test, y_train, y_test)
```

El accuracy (o precisiÃ³n) del conjunto de testeo fue de 0.951; un resultado muy alto y una precisiÃ³n muy alta. Lo cual indica que nuestro modelo es capaz de predecir resultado con mucha precisiÃ³n y es confiable.

# Clasificador Bagging

```python
bag_clf = BaggingClassifier(random_state=42)
bag_clf.fit(X_train, y_train)

evaluate_classification(bag_clf, X_train, X_test, y_train, y_test)
```

El accuracy del modelo de Bagging tambiÃ©n resultÃ³ ser muy alto, arrojando 0.951 y tambiÃ©n es capaz de predecir resultados con mucha precisiÃ³n y ser confiable.

# Random Forest

```python
rf_clf = RandomForestClassifier(random_state=42)
rf_clf.fit(X_train, y_train)

evaluate_classification(rf_clf, X_train, X_test, y_train, y_test)
```

El accuracy del modelo Random Forest arrojÃ³ un valor levemente mÃ¡s alto que el de los otros dos modelos anteriores, lo cual indica mayor precisiÃ³n en los resultados. Resulta ser mÃ¡s confiable que los otros dos, pero sin menospreciarlos.

# Matriz de ConfusiÃ³n

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(rf_clf, X_test, y_test, cmap = 'Blues');
```
